PREDICTION DE L'INSUFFISANCE CARDIAQUE

Qu'est-ce que l'insuffisance cardiaque ?

On parle d'insuffisance cardiaque lorsque le cœur perd une partie de sa force musculaire et sa capacité de contraction normale ; il ne pompe plus suffisamment de sang pour permettre aux organes de recevoir assez d'oxygène et d'éléments nutritifs, essentiels à leur bon fonctionnement.

Auteur: PARFAIT TANOH N'GORAN

Attributs de l'ensemble de données

Age : âge [années]

anémie : Diminution des globules rouges ou de l'hémoglobine (booléen)

creatinine_phosphokinase : Niveau de l'enzyme CPK dans le sang (mcg/L)

diabetes : Si le patient est diabétique (booléen)

ejection_fraction : Pourcentage de sang sortant du cœur à chaque contraction (pourcentage)

high_blood_pressure : Si le patient souffre d'hypertension

(booléen) platelets : Plaquettes dans le sang (kiloplatelets/mL)

serum_creatinine : Niveau de créatinine sérique dans le sang

(mg/dL) serum_sodium : Taux de sodium sérique dans le sang
(mEq/L)

sexe : Femme ou homme (binaire)

smoking : Si le patient fume ou non (booléen) 


time: Période de suivi (jours)

DEATH_EVENT : Si le patient est décédé pendant la période de suivi (booléen

```python
def catplot(df,x):
    sns.set(style='white')
    fig = plt.subplots(1,3,figsize = (15,4))
    plt.subplot(1,3,1)
    df[x].value_counts().plot.pie(explode=[0.1,0.1], autopct='%1.1f%%', shadow=True)
    plt.subplot(1,3,2)
    ax=sns.histplot(data=df,x=x,kde = True,color=couleurs[0],edgecolor = 'k')
    ax.bar_label(ax.containers[0])
    plt.subplot(1,3,3)
    ax=sns.countplot(data=df, x=x, hue='Status',palette = couleurs,edgecolor = 'k')
    for container in ax.containers:
        ax.bar_label(container)
        tit = x + ' vs Événement de décès'
        plt.suptitle(tit)
```

```python
def numplot(df,x,scale):
    sns.set(style='whitegrid')
    fig = plt.subplots(2,1,figsize = (15,11))
    plt.subplot(2,1,1)
    ax=sns.histplot(data=df, x=x, kde=True,color=couleurs[0],edgecolor = 'k')
    ax.bar_label(ax.containers[0])
    tit=x + ' distribution'
    plt.title(tit)
    plt.subplot(2,1,2)
    tar=x + '_group'
    Tstr= str(scale)
    tit2=x + ' vs Événement de décès ( ' + Tstr + ' : 1 )'
    df[tar] = [ int(i / scale) for i in df[x]]
    ax=sns.countplot(data=df, x=tar, hue='Status',palette = couleurs,edgecolor = 'k')
    for container in ax.containers:
        ax.bar_label(container)
        plt.title(tit2)
        
```

Aperçu des caractéristiques numériques :
Les cas de DEATH_EVENT commencent à l'âge de 45 ans. Certains pics spécifiques de cas
élevés de DEATH_EVENT peuvent être observés à 45, 50, 60, 65 et 70 ans.
Des cas élevés de décès peuvent être observés pour des valeurs de fraction d'éjaculation
comprises entre 20 et 60. Des valeurs de créatinine sérique comprises entre 0,6 et 3,0 ont
une probabilité plus élevée de conduire à un décès. Des valeurs de sodium sérique
comprises entre 127 et 145 indiquent un décès dû à une insuffisance cardiaque.
Les cas de DEATH_EVENT sont élevés pour les valeurs comprises entre 0(0x100) - 500(5x100) pour la
créatinine_phosphokinase.
les valeurs de plaquettes comprises entre 0(0x10^5) et 400 000(4x10^5) sont sujettes à des défaillances
cardiaques entraînant DEATH_EVENT.
Pour la caractéristique de temps, les valeurs comprises entre 0(0x10) et 60(6*10) ont une probabilité plus
élevée de conduire à un DEATH_EVENT.
Résumé des caractéristiques catégorielles :
âge : 50 - 70 ans
créatinine_phosphokinase : 0 - 500
Fraction de l'action : 20 - 40
plaquettes : 200 000 - 300 000
temps : 0 - 50

```python
y = df['DEATH_EVENT']
x = df.drop(['DEATH_EVENT'], axis = 1)
```

TRAITEMENT DES DONNEES DE DESEQUILIBRE

```python
x_resample, y_resample = SMOTE().fit_resample(x, y)
```

```python
print(x_resample.shape)
print(y_resample.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_resample, y_resample, test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
def mymodel(model):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    train_accuracy = model.score(x_train,y_train)
    test_accuracy = model.score(x_test,y_test)
    print("Model :-",str(model))
    print('\nModel Accuracy: ', accuracy_score(y_test, y_pred))
    print(f'\nTraining Accuracy:{train_accuracy} \nTesting Accuracy :{test_accuracy}')
    print('--------------------------------------')
    print()
    return model
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
knn = mymodel(KNeighborsClassifier())
svc = mymodel(SVC())
dt = mymodel(DecisionTreeClassifier())
lr = mymodel(LogisticRegression())
gnb = mymodel(GaussianNB())
rfc = mymodel(RandomForestClassifier())
xgb = mymodel(XGBClassifier())
```
